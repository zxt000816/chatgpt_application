```python
from langchain.embeddings import OpenAIEmbeddings
```

```python
embeddings = OpenAIEmbeddings()
```

```python
text = "This is a test document."
```

```python
query_result = embeddings.embed_query(text)
print(query_result)
```

```python
doc_result = embeddings.embed_documents([text])
print(doc_result)
```
